```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = '/data/bathir/ANS/R_ANS/MD/Analysis/Distances/ANS_R_distances.dat'  # replace with your .dat file path
with open(file_path, 'r') as file:
    first_line = file.readline().strip()
column_names = first_line.split()  # Assuming columns are separated by whitespace

# Load the rest of the data into a DataFrame, skipping the first line
data = pd.read_csv(file_path, sep='\s+', names=column_names, skiprows=1)

# Print column names to verify
print("Column names:", column_names)
```

```python
# Assume the distances are in the first column and angles in the second column (adjust as needed)
distance_column = "R-Chiral_H"
angle_column = "Angle-R-Chiral_H"

# Plot scatter plot for distance vs angle
plt.figure(figsize=(10, 6))
plt.scatter(data[angle_column], data[distance_column],
            alpha=0.7, edgecolors='k')
plt.title('Scatter Plot of Distance vs Angle')
plt.xlabel('Angle')
plt.ylabel('Distance')
plt.grid(True)
plt.show()

```
