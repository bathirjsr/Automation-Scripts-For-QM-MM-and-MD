```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import groupby

# Load the DCCA matrix from the DCCA.dat file
dcca_matrix = np.loadtxt(
    '/data/bathir/EFE/EFE_WT/EFE-Bollinger-Mechanism/A198L/MD/Analysis/DCCA/DCCA.dat')

# Define a threshold for high correlation and anticorrelation
high_corr_threshold = 0.3
high_anticorr_threshold = -0.3

# Function to find contiguous ranges


def find_ranges(indices):
    ranges = []
    for k, g in groupby(enumerate(indices), lambda x: x[1] - x[0]):
        group = list(g)
        ranges.append((group[0][1], group[-1][1]))
    return ranges


# Initialize dictionaries to store ranges
correlated_ranges = {}
anticorrelated_ranges = {}

# Find ranges of highly correlated and anticorrelated residues for each residue
for i in range(dcca_matrix.shape[0]):
    correlated_indices = np.where(dcca_matrix[i] > high_corr_threshold)[0]
    anticorrelated_indices = np.where(
        dcca_matrix[i] < high_anticorr_threshold)[0]

    correlated_ranges[i+1] = find_ranges(correlated_indices)
    anticorrelated_ranges[i+1] = find_ranges(anticorrelated_indices)

# Print the correlated and anticorrelated ranges for each residue
print("Highly correlated ranges:")
for res, ranges in correlated_ranges.items():
    if ranges:
        range_str = ', '.join(
            [f"{start+1}-{end+1}" if start != end else f"{start+1}" for start, end in ranges])
        print(f"Residue {res}: {range_str}")

print("\nHighly anticorrelated ranges:")
for res, ranges in anticorrelated_ranges.items():
    if ranges:
        range_str = ', '.join(
            [f"{start+1}-{end+1}" if start != end else f"{start+1}" for start, end in ranges])
        print(f"Residue {res}: {range_str}")
```
